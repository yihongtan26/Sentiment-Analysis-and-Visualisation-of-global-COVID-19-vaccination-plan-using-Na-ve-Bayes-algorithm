```python
import twint
import nest_asyncio
nest_asyncio.apply()


def twintCrawler(keywords, language,since,until):
    i = 0
    for keyword in keywords:
        c = twint.Config()
        c.Limit = 500
        c.Since = since
        c.Until = until
        c.Lang = language
        c.Search = [keyword]
        c.Pandas = True
        c.Stats = False
        c.Hide_output = True
        c.Store_csv = True
        c.extend = True;
        c.Output = "finaltest.csv"
        twint.run.Search(c)
        i+=1
        print(i, ' ', keyword, ' Done Crawling !')
        if(i==3):
            print('=========================================================================================================')

keywords = ['covid-19 vaccine','covid-19 vaccination','covid vaccine']
```

```python
twintCrawler(keywords,"en","2021-01-01","2021-01-15")
```

```python
from datetime import date, timedelta

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2020, 12, 31)
end_date = date(2021, 7, 31)
for single_date in daterange(start_date, end_date):
    next_date = single_date + timedelta(days=1)
    single_date = single_date.strftime("%Y-%m-%d")
    next_date = next_date.strftime("%Y-%m-%d")
    twintCrawler(keywords, "en", single_date, next_date)
```
